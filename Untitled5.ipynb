图片和lable变成tensor，分train和val，并设置好如何load data（每次自动提取一个batch的tensor
）

```python
import torch
from torch.utils.data import Dataset, DataLoader, random_split
from torchvision import transforms
import torch.nn as nn
```

```python
class EyeDataset(Dataset):
  def __init__ (self,csv_path,img_dir):
    self.df=pd.read_csv(csv_path)
    self.img_dir=Path(img_dir)
    self.transform=transforms.ToTensor()

  def __len__(self):
    return len(self.df)

  def __getitem__ (self,idx):
    row=self.df.iloc[idx]
    img_path=self.img_dir/row["image"]
    img = Image.open(img_path)
    w, h = img.size
    img = img.resize((int(w*0.3), int(h*0.3)))

    img_tensor=self.transform(img)

    label_tensor=torch.tensor (
        [row["x1"],row["y1"],row["x2"],row["y2"]],
        dtype=torch.float32
        )
    return img_tensor, label_tensor

csv_path="/content/data/lable/label_pic.csv"
img_gr_dir="/content/data/gray_pic"

full_dataset= EyeDataset(csv_path,img_gr_dir)
print("Total:", len(full_dataset))

train_size=int(0.8*len(full_dataset))
val_size=len(full_dataset)-train_size

generator = torch.Generator().manual_seed(42)

train_dataset,val_dataset=random_split(
    full_dataset,
     [train_size,val_size],
    generator=generator
    )

train_loader=DataLoader(train_dataset,batch_size=4,shuffle=True)
val_loader=DataLoader(val_dataset,batch_size=32,shuffle=False)

images, labels = next(iter(train_loader))

print("images.shape =", images.shape)
print("labels.shape =", labels.shape)
print("labels example =")
print(labels)
```

创建CNN模型，用于确定flatten数据的方法，用AdaptiveAvgPool2d把每张图的特征确定为1，1

```python
class EyeCNN(nn.Module):
  def __init__(self):
    super().__init__()

    self.features=nn.Sequential(
        nn.Conv2d(1,8,3,padding=1),
        nn.ReLU(),
        nn.MaxPool2d(2),

        nn.Conv2d(8,16,3,padding=1),
        nn.ReLU(),
        nn.MaxPool2d(2),

        nn.Conv2d(16,32,3,padding=1),
        nn.ReLU(),
        nn.MaxPool2d(2),

        nn.Conv2d(32,32,3,padding=1),
        nn.ReLU(),
        nn.MaxPool2d(2)
    )

    self.pool=nn.AdaptiveAvgPool2d((1,1))
    self.fc = nn.Linear(32,4)

  def forward(self,x):
    x=self.features(x)
    x=self.pool(x)
    x=torch.flatten(x,1)
    x=self.fc(x)
    return x
```

创建模型。损失函数，优化器

```python
model = EyeCNN()
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0008)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='min',
    factor=0.2,
    patience=3
)
```

```python
epochs=50

train_losses = []
val_losses = []

for epoch in range(epochs):

  model.train()

  train_loss=0.0

  for images,lables in train_loader:
    preds=model(images)
    loss=criterion(preds,lables)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    train_loss+=loss.item()

  avg_train_loss=train_loss/len(train_loader)
  scheduler.step(avg_train_loss)



  model.eval()
  val_loss=0.0

  with torch.no_grad():
    for images,labels in val_loader:
      preds=model(images)
      loss=criterion(preds,labels)
      val_loss+=loss.item()

  avg_val_loss=val_loss/len(val_loader)

  train_losses.append(avg_train_loss)
  val_losses.append(avg_val_loss)



  print(f"Epoch{epoch+1}/{epochs}, train loss:{avg_train_loss:6f},val loss:{avg_val_loss:6f}，lr:{optimizer.param_groups[0]['lr']:.6f}")
```

```python
import matplotlib.pyplot as plt
from datetime import datetime

result_name=f"1 pic size:ori, batch size:4, f{epochs}, cnn:4c4n，la:0.0003,ReduceLROnPlateau"
exp_dir=Path("/content/result")
exp_dir.mkdir(parents=True,exist_ok=True)
time_str = datetime.now().strftime("%Y%m%d_%H%M%S")

plt.figure()

plt.plot(train_losses,label="train loss")
plt.plot(val_losses,label="val loss")

plt.xlabel("Epoch")
plt.ylabel("loss")
plt.title("training Curve")
plt.legend()

result_path=exp_dir/f"{result_name}{time_str}.png"
plt.savefig(result_path)
plt.show()

print("图已保存:", result_path)
```

```python
pre_dir=Path("/content/result/pre vs label")
pre_dir.mkdir(parents=True,exist_ok=True)

model.eval()
all_images = []
all_labels = []
all_preds = []
with torch.no_grad():
    for images, labels in val_loader:

        preds = model(images)

        all_images.append(images)
        all_labels.append(labels)
        all_preds.append(preds)

images = torch.cat(all_images)
labels = torch.cat(all_labels)
preds = torch.cat(all_preds)

for i in range(len(images)):
  img=images[i].squeeze().cpu().numpy()
  true = labels[i].cpu().numpy()
  pred = preds[i].cpu().numpy().clip(0,1)

  H, W = img.shape

  tx1, ty1 = true[0]*W, true[1]*H
  tx2, ty2 = true[2]*W, true[3]*H

  px1, py1 = pred[0]*W, pred[1]*H
  px2, py2 = pred[2]*W, pred[3]*H

  plt.figure()
  plt.imshow(img, cmap="gray")

  plt.scatter(tx1, ty1, c="green")
  plt.scatter(tx2, ty2, c="green")

  plt.scatter(px1, py1, c="red")
  plt.scatter(px2, py2, c="red")

  save_path = os.path.join(pre_dir, f"sample_{i}.png")
  plt.savefig(save_path)

  plt.show()

```

总体：先把图片转灰度并转成 image tensor，用torhc自带的dataset和dataloader，写一个逐步把图片和标签变成tsensor并逐步喂入的过程，其中包括分训练集和val。然后写cnn和损失函数。训练时只把 image 输入 CNN，模型输出 4 个归一化坐标；再用输出和真实 label 算损失。然后用 zero_grad()、backward()、step() 完成一次参数更新。验证时只前向传播和算 loss，不更新参数。最后把预测的归一化坐标乘回图片宽高，在图上画出来肉眼检查
