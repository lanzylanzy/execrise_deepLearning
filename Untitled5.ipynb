standard the size of pic

```python
from pathlib import Path
from PIL import Image

input_dir=Path("/content/sample_data/raw_pic")
output_dir=Path("/content/sample_data/standard_pic")

img_path = list(input_dir.glob("*"))

max_w=0
max_h=0

for p in img_path:
  with Image.open(p) as img:
    w,h=img.size
    max_w=max(max_w,w)
    max_h=max(max_h,h)

output_dir.mkdir(parents=True, exist_ok=True)

for p in img_path:
  with Image.open(p) as img:
    w,h =img.size
    canvas=Image.new("RGB",(max_w,max_h),(0,0,0))

    x_offest=(max_w-w)//2
    y_offest=(max_h-h)//2

    canvas.paste(img,(x_offest,y_offest))

    out_name=p.stem+".png"
    out_path=output_dir/out_name
    canvas.save(out_path, format="PNG")
```
